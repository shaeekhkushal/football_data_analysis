```python
import matplotlib.pyplot as plt
import numpy as np

# Filter only Premier League data
epl_df = df[df["Comp"] == "eng Premier League"].copy()

# Ensure required columns are numeric
for col in ["Gls", "Ast", "G+A_90", "xG", "xAG"]:
    epl_df[col] = pd.to_numeric(epl_df[col], errors="coerce")

# Add total cards column if not already added
if "Total_Cards" not in epl_df.columns:
    epl_df["Total_Cards"] = epl_df["CrdY"] + epl_df["CrdR"]

# Aggregate per club
epl_stats = epl_df.groupby("Squad").agg({
    "Gls": "sum",
    "Ast": "sum",
    "Total_Cards": "sum",
    "G+A_90": "mean",
    "xG": "mean",
    "xAG": "mean"
}).sort_values(by="Gls", ascending=False).head(20)

# Normalize for radar
metrics = epl_stats.columns.tolist()
angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False).tolist()
angles += angles[:1]

# Radar Plot
plt.figure(figsize=(8, 8))
for idx, row in epl_stats.iterrows():
    values = row.values
    values = values / np.max(epl_stats.values, axis=0)  # normalize across teams
    values = np.concatenate((values, [values[0]]))  # close the loop
    plt.polar(angles, values, label=idx, linewidth=2)
    plt.fill(angles, values, alpha=0.1)

plt.xticks(angles[:-1], metrics, fontsize=11)
plt.title("Top EPL Clubs Radar (Based on Key Metrics)", fontsize=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.4, 1.1))
plt.tight_layout()
plt.savefig("epl_club_radar.png")
plt.show()
```

```python
# üìå Filter only EPL players
epl_df = df[df["Comp"] == "eng Premier League"].copy()

# üîß Convert necessary columns to numeric
cols = ["Gls", "Ast", "xG", "xAG", "MP", "Min", "CrdY", "CrdR", "G+A_90", "G-PK"]
for col in cols:
    epl_df[col] = pd.to_numeric(epl_df[col], errors="coerce")

# üßÆ Create derived metrics
epl_df["Total_Cards"] = epl_df["CrdY"] + epl_df["CrdR"]
epl_df["G+A"] = epl_df["Gls"] + epl_df["Ast"]

# üéØ Top 2 Goalkeepers (most matches, lowest cards)
gk_mask = epl_df["Pos"].str.contains("GK", na=False)
top_gk = epl_df[gk_mask].sort_values(by=["MP", "Total_Cards"], ascending=[False, True]).drop_duplicates("Player").head(2)

# üõ°Ô∏è Top 5 Defenders (based on Min + G+A)
df_mask = epl_df["Pos"].str.contains("DF", na=False)
top_df = epl_df[df_mask].sort_values(by=["Min", "G+A"], ascending=[False, False]).drop_duplicates("Player").head(5)

# üé© Top 5 Midfielders (based on G+A and xAG)
mf_mask = epl_df["Pos"].str.contains("MF", na=False)
top_mf = epl_df[mf_mask].sort_values(by=["G+A", "xAG"], ascending=[False, False]).drop_duplicates("Player").head(5)

# ‚öΩ Top 3 Forwards (based on goals and xG)
fw_mask = epl_df["Pos"].str.contains("FW", na=False)
top_fw = epl_df[fw_mask].sort_values(by=["Gls", "xG"], ascending=[False, False]).drop_duplicates("Player").head(3)

# üß© Combine all positions into final 15-man squad
squad_df = pd.concat([top_gk, top_df, top_mf, top_fw])
squad_df = squad_df[["Player", "Squad", "Pos", "MP", "Min", "Gls", "Ast", "G+A", "xG", "xAG", "Total_Cards"]]

# üíæ Optional CSV Export
squad_df.to_csv("epl_best_squad_24_25.csv", index=False)

# üìã Display result
print("‚úÖ Performance-Based 15-Man EPL Squad (2024/25):")
display(squad_df)
```

```python
import matplotlib.pyplot as plt

# ------- helper ------- #
def assign_roles(squad):
    roles = {"GK": [], "DF": [], "MF": [], "FW": [], "SUB": []}
    for _, r in squad.iterrows():
        pos = r["Pos"]
        if "GK" in pos and len(roles["GK"]) < 1:
            roles["GK"].append(r)
        elif "DF" in pos and len(roles["DF"]) < 4:
            roles["DF"].append(r)
        elif "MF" in pos and len(roles["MF"]) < 3:
            roles["MF"].append(r)
        elif "FW" in pos and len(roles["FW"]) < 3:
            roles["FW"].append(r)
        else:
            roles["SUB"].append(r)
    return roles
# ---------------------- #

roles = assign_roles(squad_df)

# pitch setup
fig, ax = plt.subplots(figsize=(14, 9))
pitch_color = "#006700"
fig.patch.set_facecolor(pitch_color)
ax.set_facecolor(pitch_color)
ax.set_xlim(0, 100); ax.set_ylim(0, 110)
ax.axis("off")

# simple white outline
ax.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], color="white", lw=2)

# 4‚Äë3‚Äë3 coordinates
coords = {
    "GK": [(50, 10)],
    "DF": [(20, 30), (40, 30), (60, 30), (80, 30)],
    "MF": [(30, 55), (50, 60), (70, 55)],
    "FW": [(25, 80), (50, 85), (75, 80)],
}

# plot starters
for role, pts in coords.items():
    for i, (x, y) in enumerate(pts):
        if i < len(roles[role]):
            p = roles[role][i]
            ax.scatter(x, y, s=600, color="gold", edgecolors="black", zorder=3)
            ax.text(x, y + 4, p["Player"], ha="center", va="bottom",
                    color="white", fontsize=16, weight="bold", zorder=4)

# üß© Display substitutes below pitch (centered)
subs = roles["SUB"]
n_subs = len(subs)
x_start = 50 - (n_subs * 9) / 2
for idx, sub in enumerate(subs):
    ax.text(x_start + idx * 13, 103, f"SUB: {sub['Player']}",
            ha="center", va="bottom",
            color="white", fontsize=10, fontweight="bold",
            bbox=dict(fc="#555", ec="white", boxstyle="round,pad=0.5"))

# üè∑Ô∏è title
plt.title("Best EPL Squad Prediction for 2025/26 (4 3 3)", color="white", fontsize=16, weight="bold")
plt.tight_layout()
plt.savefig("epl_pitch_squad.png")
plt.show()
```
